```python
import pandas as pd 

Files = ["DataFiles\JC-202006-citibike-tripdata.csv",
         "DataFiles\JC-202005-citibike-tripdata.csv",
         "DataFiles\JC-202004-citibike-tripdata.csv",
         "DataFiles\JC-202003-citibike-tripdata.csv",
         "DataFiles\JC-202002-citibike-tripdata.csv",
         "DataFiles\JC-202001-citibike-tripdata.csv",
         "DataFiles\JC-201912-citibike-tripdata.csv",
         "DataFiles\JC-201911-citibike-tripdata.csv",
         "DataFiles\JC-201910-citibike-tripdata.csv",
         "DataFiles\JC-201909-citibike-tripdata.csv",
         "DataFiles\JC-201908-citibike-tripdata.csv",
         "DataFiles\JC-201907-citibike-tripdata.csv",
         "DataFiles\JC-201906-citibike-tripdata.csv",
         "DataFiles\JC-201905-citibike-tripdata.csv",
         "DataFiles\JC-201904-citibike-tripdata.csv",
         "DataFiles\JC-201903-citibike-tripdata.csv",
         "DataFiles\JC-201902-citibike-tripdata.csv",
         "DataFiles\JC-201901-citibike-tripdata.csv",
         "DataFiles\JC-201812-citibike-tripdata.csv",
         "DataFiles\JC-201811-citibike-tripdata.csv",
         "DataFiles\JC-201810-citibike-tripdata.csv",
         "DataFiles\JC-201809-citibike-tripdata.csv",
         "DataFiles\JC-201808-citibike-tripdata.csv",
         "DataFiles\JC-201807-citibike-tripdata.csv",
         "DataFiles\JC-201806-citibike-tripdata.csv",
         "DataFiles\JC-201805-citibike-tripdata.csv",
         "DataFiles\JC-201804-citibike-tripdata.csv",
         "DataFiles\JC-201803-citibike-tripdata.csv",
         "DataFiles\JC-201802-citibike-tripdata.csv",
         "DataFiles\JC-201801-citibike-tripdata.csv",
         "DataFiles\\NY-202006-citibike-tripdata.csv",
         "DataFiles\\NY-202005-citibike-tripdata.csv",
         "DataFiles\\NY-202004-citibike-tripdata.csv",
         "DataFiles\\NY-202003-citibike-tripdata.csv",
         "DataFiles\\NY-202002-citibike-tripdata.csv",
         "DataFiles\\NY-202001-citibike-tripdata.csv",
         "DataFiles\\NY-201912-citibike-tripdata.csv",
         "DataFiles\\NY-201911-citibike-tripdata.csv",
         "DataFiles\\NY-201910-citibike-tripdata.csv",
         "DataFiles\\NY-201909-citibike-tripdata.csv",
         "DataFiles\\NY-201908-citibike-tripdata.csv",
         "DataFiles\\NY-201907-citibike-tripdata.csv",
         "DataFiles\\NY-201906-citibike-tripdata.csv",
         "DataFiles\\NY-201905-citibike-tripdata.csv",
         "DataFiles\\NY-201904-citibike-tripdata.csv",
         "DataFiles\\NY-201903-citibike-tripdata.csv",
         "DataFiles\\NY-201902-citibike-tripdata.csv",
         "DataFiles\\NY-201901-citibike-tripdata.csv",
         "DataFiles\\NY-201812-citibike-tripdata.csv",
         "DataFiles\\NY-201811-citibike-tripdata.csv",
         "DataFiles\\NY-201810-citibike-tripdata.csv",
         "DataFiles\\NY-201809-citibike-tripdata.csv",
         "DataFiles\\NY-201808-citibike-tripdata.csv",
         "DataFiles\\NY-201807-citibike-tripdata.csv",
         "DataFiles\\NY-201806-citibike-tripdata.csv",
         "DataFiles\\NY-201805-citibike-tripdata.csv",
         "DataFiles\\NY-201804-citibike-tripdata.csv",
         "DataFiles\\NY-201803-citibike-tripdata.csv",
         "DataFiles\\NY-201802-citibike-tripdata.csv",
         "DataFiles\\NY-201801-citibike-tripdata.csv"]

df_ALLStations = pd.DataFrame()

for file in Files:
    print(file)
    dataStations = pd.read_csv(file)
    
    # Removed trips more than 1 day
    dataStations = dataStations.loc[dataStations["tripduration"] < 86400]

    df_Stations = dataStations[["start station id","start station name","start station latitude","start station longitude"]].drop_duplicates()
    df_Stations.columns = ["station id","station name","latitude","longitude"]

    df_Stations2 = dataStations[["end station id","end station name","end station latitude","end station longitude"]].drop_duplicates()
    df_Stations2.columns = ["station id","station name","latitude","longitude"]

    df_ALLStations = df_ALLStations.append(df_Stations)
    df_ALLStations = df_ALLStations.append(df_Stations2).drop_duplicates()

    df_ALLStations = df_ALLStations.drop_duplicates(subset=['station id'], keep='first')

df_ALLStations.sort_values("station id",inplace=True)
df_ALLStations = df_ALLStations.drop_duplicates(subset=['station id'], keep='first')
```
